```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_sql("SELECT * FROM Ans", mydb)
df
```

```python
Avg_Alcohol_content = pd.read_sql("SELECT AVG(Alcohol) AS Average_Alcohol_Count FROM Ans", mydb)
Avg_Alcohol_content
```

```python
# Find the wines with the highest Alcohol content.
# Subquery: Select wines where alcohol is greater than the average alcohol content.
Wine = pd.read_sql("SELECT * FROM Ans WHERE Alcohol > (SELECT AVG(Alcohol) FROM Ans)", mydb)
Wine
```

```python
# List wines with a Proline value greater than the average.
# Subquery: Compare Proline value to the average in the dataset.
Proline_Value = pd.read_sql("SELECT * FROM Ans WHERE Proline > (SELECT AVG(Proline) FROM Ans)", mydb)
Proline_Value
```

```python
# Find wines with higher Flavanoids than the wine with the lowest Malic_Acid.
# Subquery: Get the minimum Malic_Acid value and compare Flavanoids.
Minim_Malic_Acid = pd.read_sql(
    """
    SELECT * 
    FROM Ans 
    WHERE Flavanoids > (
        SELECT Flavanoids 
        FROM Ans 
        WHERE Malic_Acid = (SELECT MIN(Malic_Acid) FROM Ans)
        LIMIT 1
    )
    """, mydb
)

Minim_Malic_Acid
```

```python
# Display wines that have an Ash value higher than the average ash in the YES customer segment.
# Subquery: Filter by customer segment and get the average.
Ash_value = pd.read_sql("SELECT * FROM Ans WHERE Ash > (SELECT AVG(ASH) FROM Ans WHERE Customer_Segment='YES')", mydb)
Ash_value
```

```python
# Find wines where the Color_Intensity is greater than the average intensity of NO customer segment wines.
# Subquery: Average color intensity by segment.
Color_Intensity = pd.read_sql("SELECT * FROM Ans WHERE Color_Intensity > (SELECT AVG(Color_Intensity) FROM Ans WHERE Customer_Segment = 'NO')", mydb)
Color_Intensity
```

```python
# Select wines with Magnesium above the maximum Magnesium found in wines with below-average Alcohol.
# Subquery: Use a subquery to find maximum Magnesium where Alcohol is below the average.
Max_Magenesium = pd.read_sql("SELECT * FROM Ans WHERE Magnesium > (SELECT MAX(Magnesium) FROM Ans WHERE Alcohol < (SELECT AVG(Alcohol) FROM Ans))", mydb)
Max_Magenesium
```
