```python
import mysql.connector as connection

try:
    mydb = connection.connect(host="localhost", user="root", passwd="", database="ipl", use_pure=True)
    
    if mydb.is_connected():
        print("Successfully connected to the demo database.")
except Exception as e:
    print("Error:", str(e))
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_sql("SELECT * FROM orders;", mydb)
df
```

```python
# How many orders were placed by each customer?
Order_placed_customers = pd.read_sql("SELECT  CustomerName, COUNT(*) AS Order_count FROM orders GROUP BY CustomerName ORDER BY Order_count DESC",mydb)
Order_placed_customers
```

```python
# How many orders were placed in each state?
Order_placed_state = pd.read_sql("SELECT State, COUNT(*) Orders_state_count FROM Orders GROUP BY State ORDER BY Orders_state_count DESC", mydb)
Order_placed_state
```

```python
# How many orders were placed in each city?
Order_placed_city = pd.read_sql("SELECT City, COUNT(*) AS Order_count_city FROM Orders GROUP BY City ORDER BY Order_count_city", mydb)
Order_placed_city
```

```python
# How many unique customers are there in each city?
Distinct_customers = pd.read_sql("SELECT City, COUNT(DISTINCT CustomerName) AS Distinct_Customer_count FROM Orders GROUP BY City", mydb)
Distinct_customers
```

```python
# How many orders were placed each month?
Order_Placed_each_month = pd.read_sql("SELECT `Order Date`, COUNT(*) AS Each_month_orders FROM Orders GROUP BY `Order Date` ORDER BY Each_month_orders DESC", mydb)
Order_Placed_each_month
```

```python
Each_Month = pd.read_sql("SELECT MONTH(`Order Date`) AS Order_Month, COUNT(*) AS Order_count FROM Orders GROUP BY MONTH(`Order Date`) ORDER BY Order_count DESC", mydb)
Each_Month
```

```python
Distinct_customers_state = pd.read_sql("SELECT State, COUNT(DISTINCT CustomerName) AS Customer_count FROM Orders GROUP BY State ", mydb)
Distinct_customers_state
```

```python
Grouped_city = pd.read_sql("SELECT State, City , COUNT(*) AS City_count FROM Orders GROUP BY State, City", mydb)
Grouped_city
```

```python
ANS = pd.read_sql("SELECT EXTRACT(MONTH FROM `Order Date`) AS Month, COUNT(*) AS TotalOrders FROM Orders GROUP BY Month", mydb)
ANS
```
