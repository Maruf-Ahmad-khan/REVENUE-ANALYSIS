# Read the table

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_sql("SELECT * FROM users;", mydb)
df
```

# Write a query to find the total number of users in the dataset

```python
df_count = pd.read_sql("SELECT COUNT(*) AS Total_number_of_users FROM users;", mydb)
df_count
```

# How would you retrieve all unique states present in the dataset?

```python
df_unique_state = pd.read_sql("SELECT DISTINCT state AS Unique_state FROM users", mydb)
df_unique_state
```

# Write a SQL query to count how many users are in each state.

```python
user_count = pd.read_sql("SELECT state, COUNT(*) AS user_count FROM users GROUP BY city ORDER BY user_count ASC", mydb)
user_count
```

# How can you find the number of users in a specific city, for example, Ahmedabad?

```python
User_count_Ahmedabad = pd.read_sql("SELECT COUNT(*) AS user_count FROM users WHERE city = 'Ahmedabad';", mydb)
User_count_Ahmedabad
```

# Write a query to find the number of users in each city within a specific state, such as Maharashtra.

```python
No_of_users_Maharashtra = pd.read_sql("SELECT COUNT(*) user_count FROM users WHERE state = 'Maharashtra';", mydb)
No_of_users_Maharashtra
```

# How would you retrieve all users whose names start with the letter 'B'?

```python
users_Name_Startwith_B = pd.read_sql("SELECT * FROM users WHERE name LIKE 'b%'", mydb)
users_Name_Startwith_B
```

# Write a query to list users who live in cities that start with the letter 'J'.

```python
user_city_startwith_J = pd.read_sql("SELECT * FROM users WHERE city LIKE 'j%'", mydb)
user_city_startwith_J
```

# How can you find the user with the highest user_id?

```python
# Highest_user_id = pd.read_sql("SELECT MAX(user_id) AS Highest_user_id FROM users", mydb)
Highest_user_id = pd.read_sql("SELECT * FROM users ORDER BY user_id LIMIT 1", mydb)
Highest_user_id
```

# Write a query to return users from the state 'Gujarat' and the city 'Ahmedabad'.

```python
Gujarat_Ahmedabad = pd.read_sql("SELECT * FROM users WHERE state = 'Gujarat' AND city = 'Ahmedabad'", mydb)
Gujarat_Ahmedabad
```

# How would you get the count of users grouped by state and then by city?

```python
users_state_city = pd.read_sql("SELECT state, city, COUNT(*) AS user_count FROM users GROUP BY state ,city ORDER BY state, city DESC;", mydb)
users_state_city
     
```
