```python
df_count = pd.read_sql("SELECT COUNT(*) AS Total_number_of_users FROM users;", mydb)
df_count
```

# How would you retrieve all unique states present in the dataset?

```python
df_unique_state = pd.read_sql("SELECT DISTINCT state AS Unique_state FROM users", mydb)
df_unique_state
```

# Write a SQL query to count how many users are in each state.

```python
user_count = pd.read_sql("SELECT state, COUNT(*) AS user_count FROM users GROUP BY city ORDER BY user_count ASC", mydb)
user_count
```

# How can you find the number of users in a specific city, for example, Ahmedabad?

```python
User_count_Ahmedabad = pd.read_sql("SELECT COUNT(*) AS user_count FROM users WHERE city = 'Ahmedabad';", mydb)
User_count_Ahmedabad
```

# Write a query to find the average user_id for users in each state.

```python
Average_user_id = pd.read_sql("SELECT user_id, state, AVG(user_id) AS Avg_user_id FROM users GROUP BY state ORDER BY Avg_user_id;", mydb)
Average_user_id
```

# How would you find the city with the maximum number of users?

```python
Max_num_users = pd.read_sql("SELECT  city, COUNT(*) AS Maxm_num_users FROM users GROUP BY city ORDER BY Maxm_num_users DESC LIMIT 1", mydb)
Max_num_users
```

# Write a query to find the state with the minimum number of cities represented.

```python
Min_city_count = pd.read_sql("SELECT state, COUNT(DISTINCT city) AS Minm_num_city FROM users GROUP BY state ORDER BY Minm_num_city ASC LIMIT 1 ;", mydb)
Min_city_count
```

# How would you group users by the first letter of their city name?

```python
City_First_Letter = pd.read_sql("SELECT LEFT(city, 1) AS First_Letter, COUNT(*) AS city_count FROM users GROUP BY LEFT(city, 1) ORDER BY city_count DESC;", mydb)
City_First_Letter
```

# Write a query to find the top 5 cities with the most users.

```python
Top_5_cities = pd.read_sql("SELECT city, COUNT(*) AS Top_five_cities FROM users GROUP BY city ORDER BY Top_five_cities DESC LIMIT 5", mydb)
Top_5_cities
```

# Write a query to find all users who live in the same city as the user with user_id 1 .


```python
same_user_id = pd.read_sql(
     "SELECT user_id, name, state, city FROM users WHERE city = (SELECT city FROM users WHERE user_id = 1)"
     , mydb)
same_user_id
```

# How would you find states that have more users than the state 'Rajasthan'?

```python
More_users = pd.read_sql("SELECT state, COUNT(*) AS user_count FROM users GROUP BY state HAVING COUNT(*) > (SELECT COUNT(*) FROM users WHERE state = 'Rajasthan') ORDER BY user_count DESC", mydb)
More_users
```

# Write a query to retrieve users whose user_id is higher than the average user_id.

```python
Avg_user_id = pd.read_sql("SELECT user_id, name, state, city FROM users WHERE user_id > (SELECT AVG(user_id) FROM users)", mydb)
Avg_user_id
```

# How would you find the second most populated city in the dataset?

```python
Second_populated_city = pd.read_sql("SELECT city, COUNT(*) user_city FROM users GROUP BY city ORDER BY user_city DESC LIMIT 1 OFFSET 1", mydb)
Second_populated_city
```

# Write a query to list users who live in cities that have fewer than 5 users.

```python
Less_user = pd.read_sql("SELECT user_id, name, state, city FROM users WHERE city IN (SELECT city FROM users GROUP BY city HAVING COUNT(*) < 5)", mydb)
Less_user
```
