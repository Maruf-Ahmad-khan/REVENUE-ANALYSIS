# 1. Find the orders where the profit is higher than the average profit for all orders.

```python
Higher_Profit = pd.read_sql("SELECT * FROM Details WHERE Profit > (SELECT AVG(Profit) FROM Details)", mydb)
Higher_Profit
```

# 2. List the orders that have a quantity more than the minimum quantity ordered for the 'Furniture' category.

```python
Minm_Quantity = pd.read_sql("SELECT * FROM Details WHERE Quantity > (SELECT MIN(Quantity) FROM Details WHERE Category = 'Furniture')", mydb)
Minm_Quantity
```

# 3.  Retrieve the top 3 most profitable orders in each subcategory.

```python
Top_Most_Orders = pd.read_sql("SELECT * FROM Details WHERE Order_Id = (SELECT COUNT(Profit) FROM Details GROUP BY Sub_Category  ORDER BY Profit DESC LIMIT 1)", mydb)
Top_Most_Orders
```

# 4. Find the orders that have a profit greater than the highest profit for 'Electronic Games'.

```python
Electronic_Game = pd.read_sql("SELECT * FROM Details WHERE Profit > (SELECT MAX(Profit) FROM Details WHERE Sub_Category = 'Electronic Games')", mydb)
Electronic_Game
```

# 5. List all orders where the amount is less than the average amount for 'Electronics'.

```python
List_orders = pd.read_sql("SELECT * FROM Details WHERE Amount < (SELECT  AVG(Amount) FROM Details WHERE Category = 'Electronics') ", mydb)
List_orders
```

# 6. Retrieve the orders where the category is 'Furniture' and the payment mode is not one of the modes used for 'Phones'.

```python
Retrieve_Orders = pd.read_sql("SELECT * FROM Details WHERE Category = 'Furniture' AND Payment_Mode NOT IN  (SELECT DISTINCT Payment_Mode FROM Details WHERE Sub_Category = 'Phone')", mydb)
Retrieve_Orders
```

# 7. Find orders with a payment mode that is used by more than one category.

```python
Orders_Payment_Mode = pd.read_sql("SELECT * FROM Details WHERE Payment_Mode IN (SELECT Payment_Mode FROM Details GROUP BY Payment_Mode HAVING COUNT(DISTINCT Category) > 1)", mydb)
Orders_Payment_Mode
```

# 8. List the orders where the amount is greater than the average amount for their respective subcategory.
