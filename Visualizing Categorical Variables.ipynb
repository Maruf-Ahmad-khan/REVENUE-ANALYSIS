```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the Visualize class
class Visualize:
    def __init__(self, data):
        self.data = data

    def Visual3(self):
        # Sort the data by 'Percentage_%' in descending order
        df = self.data.sort_values(by='Percentage_%', ascending=False)

        # Calculate the cumulative percentage
        df['Cumulative_%'] = df['Percentage_%'].cumsum()

        # Plotting the Pareto chart
        fig, ax1 = plt.subplots(figsize=(10, 6))

        # Bar plot for individual percentages
        ax1.bar(df['Most_Needed_Action'], df['Percentage_%'], color='skyblue')
        ax1.set_ylabel('Percentage %', color='white')
        ax1.tick_params(axis='y', labelcolor='white')
        
        # Adjust x-axis tick labels and rotate them
        plt.xticks(rotation=45, ha='right', color='white')

        ax1.set_xlabel('Most Needed Action', color='white')
        plt.title('Pareto Chart of Most Needed Actions', color='white')

        # Plot cumulative percentage on a secondary axis
        ax2 = ax1.twinx()
        ax2.plot(df['Most_Needed_Action'], df['Cumulative_%'], color='red', marker='o', linestyle='-', linewidth=2)
        ax2.set_ylabel('Cumulative %', color='white')
        ax2.tick_params(axis='y', labelcolor='white')

        # Set the background color to black for better contrast with white labels
        fig.patch.set_facecolor('black')
        ax1.set_facecolor('black')
        ax2.set_facecolor('black')

        # Display the plot
        plt.tight_layout()
        plt.show()

# Sample data
Online_Survey = {
    'Most_Needed_Action': [
        'Improved regulation and oversight',
        'Improved transparency of financial report',
        'Improved corporate governance practice',
        'Improved enforcement of existing laws and regulations',
        'Improved market trading rules on transparency',
        'Improved auditing practices and standards'
    ],
    'Percentage_%': [29, 21, 17, 16, 11, 6]
}

# Create DataFrame
data = pd.DataFrame(Online_Survey)

# Instantiate the Visualize class with the data and call the Visual3 method
visualize = Visualize(data)
visualize.Visual3()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the BarGraphVisualizer class
class BarGraphVisualizer:
    def __init__(self, data):
        self.data = data

    def plot_bar_graph(self):
        # Create the bar graph
        fig, ax = plt.subplots(figsize=(10, 6))

        # Bar plot for individual percentages
        ax.bar(self.data['Most_Needed_Action'], self.data['Percentage_%'], color='skyblue')

        # Set white labels and a black background
        ax.set_ylabel('Percentage %', color='white')
        ax.set_xlabel('Most Needed Action', color='white')
        plt.title('Bar Graph with Black Background', color='white')
        plt.xticks(rotation=45, ha='right', color='white')
        ax.tick_params(axis='y', labelcolor='white')

        # Set the background color to black
        fig.patch.set_facecolor('black')
        ax.set_facecolor('black')

        # Display the plot
        plt.tight_layout()
        plt.show()

# Sample data
Online_Survey = {
    'Most_Needed_Action': [
        'Improved regulation and oversight',
        'Improved transparency of financial report',
        'Improved corporate governance practice',
        'Improved enforcement of existing laws and regulations',
        'Improved market trading rules on transparency',
        'Improved auditing practices and standards'
    ],
    'Percentage_%': [29, 21, 17, 16, 11, 6]
}

# Create DataFrame
data = pd.DataFrame(Online_Survey)

# Instantiate the BarGraphVisualizer class and call the plot_bar_graph method
visualizer = BarGraphVisualizer(data)
visualizer.plot_bar_graph()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a class for visualizations
class Visualize:
    def __init__(self, data):
        self.data = data

    def draw_pie_chart(self):
        # Plot the pie chart
        fig, ax = plt.subplots(figsize=(8, 8))

        # Plot the pie chart with white labels and a black background
        wedges, texts, autotexts = ax.pie(
            self.data['Percentage_%'], 
            labels=self.data['Most_Needed_Action'], 
            autopct='%1.1f%%', 
            startangle=140, 
            colors=plt.cm.Paired.colors,
            textprops={'color': 'white'}
        )

        # Set the background color to black
        fig.patch.set_facecolor('black')  # Background for the figure
        ax.set_facecolor('black')         # Background for the axes

        # Change the color of label texts to white
        for text in texts:
            text.set_color('white')
        for autotext in autotexts:
            autotext.set_color('white')

        # Set the title with white color
        plt.title('Pie Chart of Most Needed Actions by Percentage', color='white')

        # Display the plot
        plt.tight_layout()
        plt.show()

# Sample data
Online_Survey = {
    'Most_Needed_Action': [
        'Improved regulation and oversight',
        'Improved transparency of financial report',
        'Improved corporate governance practice',
        'Improved enforcement of existing laws and regulations',
        'Improved market trading rules on transparency',
        'Improved auditing practices and standards'
    ],
    'Percentage_%': [29, 21, 17, 16, 11, 6]
}

# Create DataFrame
data = pd.DataFrame(Online_Survey)

# Instantiate the Visualize class with the data and call the draw_pie_chart method
visualize = Visualize(data)
visualize.draw_pie_chart()
```
