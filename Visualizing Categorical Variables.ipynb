```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Online_Survey = {
                 'Most_Needed_Action' : ['Improved regulation and oversight', 'Improved trasnparency of financial report',
                                         'Improved corporate governance practice', 'Improved enforcement of existing laws and regulation',
                                         'Improved market trading rules on transparency', 'Improved auditing practices and standard'
                                         ],
                 'Percentage_%' : [29, 21, 17, 16, 11, 6]
}
data = pd.DataFrame(Online_Survey)
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming 'data' is your DataFrame
# Group by 'Most_Needed_Action' and sum the 'Percentage_%'
city_revenue = data.groupby('Most_Needed_Action')['Percentage_%'].sum().reset_index().sort_values(by='Percentage_%', ascending=False)

# Data for the horizontal bar chart
actions = city_revenue['Most_Needed_Action']
performance = city_revenue['Percentage_%']
y_pos = np.arange(len(actions))
error = np.random.rand(len(actions))  # Random error for illustration

fig, ax = plt.subplots()

# Create horizontal bar chart
ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(actions, color='white')  # Set y-tick labels to white
ax.invert_yaxis()  # Labels read top-to-bottom
ax.set_xlabel('Percentage (%)', color='white')
ax.set_title('Total Percentage by Most Needed Action', color='white')

# Set the x-tick labels to white
ax.tick_params(axis='x', colors='white')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming 'data' is your DataFrame
# Group by 'Most_Needed_Action' and sum the 'Percentage_%'
city_revenue = data.groupby('Most_Needed_Action')['Percentage_%'].sum().reset_index().sort_values(by='Percentage_%', ascending=False)

# Data for the bar chart (Pareto)
actions = city_revenue['Most_Needed_Action']
percentages = city_revenue['Percentage_%']

# Calculate cumulative percentage
cumulative_percentage = np.cumsum(percentages) / percentages.sum() * 100

fig, ax1 = plt.subplots()

# Create bar chart
ax1.bar(actions, percentages, color='C0')

# Set labels and title for the bar chart
ax1.set_xlabel('Most Needed Action', color='white')
ax1.set_ylabel('Percentage (%)', color='white')
ax1.set_title('Pareto Chart of Most Needed Actions', color='white')

# Set x-ticks and x-tick labels to white
ax1.tick_params(rotation = 90, axis='x', colors='white')
ax1.tick_params(axis='y', colors='white')

# Create a second y-axis for the cumulative percentage line
ax2 = ax1.twinx()
ax2.plot(actions, cumulative_percentage, color='C1', marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel('Cumulative Percentage (%)', color='white')
ax2.tick_params(axis='y', colors='white')

# Set the cumulative percentage labels to white
for i, txt in enumerate(cumulative_percentage):
    ax2.annotate(f'{txt:.1f}%', (actions[i], cumulative_percentage[i]), color='black', ha='center', va='bottom', fontweight = 'bold')

# Show grid
ax1.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.show()
```
